```python
# Importação:
from pytube import YouTube
```

```python
def audioDownload(link):
  yt = YouTube(link)# Criando o objeto youtube com a url
  audio_stream = yt.streams.filter(only_audio=True).order_by('abr').desc().first()
  print('Download executado com sucesso')
  return audio_stream.download()
```

```python
from pydub import AudioSegment
```

```python
def converter_para_wav(audio_path, formato_saida='wav'):
  '''
    Função que converte o arquivo de audio para wav.
  O arquivo que se queria utilizar pode não estar num
  formato adequado, então essa função converte o arquivo
  de audio para um formato mais adequado.
  '''
  # Carregar o arquivo de áudio
  audio = AudioSegment.from_file(audio_path)

  # Converter para WAV (formato padrão suportado pelo SpeechRecognition)
  novo_caminho = f"{audio_path.rsplit('.', 1)[0]}.{formato_saida}"
  audio.export(novo_caminho, format=formato_saida)
  return novo_caminho
```

# Processamento do Audio:

    Aqui é que a mágia acontece, aqui é o ponto principal
    do código. Eu estou usando aqui a biblioteca speech_recognition
    para transcrever o audio para texto e a biblioteca translate
    do google para traduzir o texto.

    A linguagem padrão é de inglês para português, porém tudo
    isso pode ser ajustado no código.

**NOTA:**

    Até o momento, eu não consegui transcrever videos/audio muito

```python
import speech_recognition as sr
from googletrans import Translator
```

```python
def transcrever_audio(audio_path, idioma_origem='en', idioma_destino='pt'):
  '''
    Função que transcreve e traduz o audio. Note que as entradas são
  com ingles como idioma original e português como idioma para o qual
  será traduzido, mas tudo isso pode ser alterado. Só cuidado para
  mudar de um jeito que o Google Translate entenda.
  '''
  recognizer = sr.Recognizer()

  with sr.AudioFile(audio_path) as source:
      audio_data = recognizer.record(source)

  try:
      texto_original = recognizer.recognize_google(audio_data, language=idioma_origem)

      # Tradução do texto
      translator = Translator()
      texto_traduzido = translator.translate(texto_original, dest=idioma_destino).text

      return (texto_original,texto_traduzido)

  except sr.UnknownValueError:
      print('Não foi possível transcrever o áudio')

  except sr.RequestError as e:
      print(f"Erro na requisição: {e}")
```

```python
# Antes de chamar transcrever_audio, converta o arquivo para WAV
audio_wav = converter_para_wav(audio)
```

```python
texto = transcrever_audio(audio_wav);texto[1]
```

```python
from gtts import gTTS
from IPython.display import Audio, display
```

```python
def texto_para_audio(texto, lg = 'pt'):
  '''
    Função que transforma a transcrição
  agora traduzida, em arquivo de audio.
  '''
  tts = gTTS(text=texto, lang=lg)

  # Salvar o áudio como um arquivo temporário
  arquivo_temporario = "/content/saida_audio.mp3"
  tts.save(arquivo_temporario)

  # Reproduzir o áudio no Colab
  display(Audio(arquivo_temporario, autoplay=True))
```
