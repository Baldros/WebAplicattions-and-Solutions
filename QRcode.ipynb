```python
# Importação da biblioteca
import qrcode
```

```python
# Basta inserir a URL dentro das aspas abaixo:
link = "https://github.com/Ustanague/CALIPSO/tree/main/ETL"
file_name = "qrcode.png"  # Nome do arquivo de
```

```python
# Construção do QRcode:
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data(link)
qr.make(fit=True)
img = qr.make_image(fill_color="black", back_color="white")
img.save(file_name) # salva o qrcode nos arquivos temporários
img
```
